````python
#Задание 2
import numpy as np
def nonlin(x, deriv=False):
    if deriv:
        return nonlin(x) * (1 - nonlin(x))
    return 1 / (1 + np.exp(-x))
# Создание массива данных на основе таблицы
data = np.array([
    [0.645, 33, 4, 0.34],        # Ока
    [2.880, 152, 14, 1.5],       # Газель пассажирская
    [1.110, 88, 5, 0.475],       # Лада Калина
    [10.400, 260, 3, 6],         # Камаз (3 оси)
    [2.880, 152, 3, 1.5],        # Газель грузовая
    [9.895, 122, 30, 6.2],       # ПАЗ4234 (пазик)
    [3.695, 136, 3, 3],          # ЗИЛ-5301 (бычок)
    [2.170, 116, 5, 0.6],        # УАЗ Патриот
    [27.895, 278, 150, 11]       # ЛиАЗ-6213
])
# Создаем массив выходных данных на основе предоставленной таблицы
output_data = np.array([
    [1, 0, 0],  # Ока
    [0, 1, 0],  # Газель пассажирская
    [1, 0, 0],  # Лада Калина
    [0, 0, 1],  # Камаз (3 оси)
    [0, 0, 1],  # Газель грузовая
    [0, 1, 0],  # ПАЗ4234 (пазик)
    [0, 0, 1],  # ЗИЛ-5301 (бычок)
    [1, 0, 0],  # УАЗ Патриот
    [0, 1, 0]   # ЛиАЗ-6213
])
print("Входные данные:")
print(data)

def normalize(parameter, value):
    normalization_table = {
        0: [(float('-inf'), 0.5), (0.5, 1), (1, 2), (2, 5), (5, float('inf'))],
        1: [(float('-inf'), 20), (20, 50), (50, 100), (100, 200), (200, float('inf'))],
        2: [(float('-inf'), 2), (2, 5), (5, 10), (10, 20), (20, float('inf'))],
        3: [(float('-inf'), 1), (1, 2), (2, 3), (3, 4), (4, float('inf'))]
    }

    for interval_index, interval in enumerate(normalization_table[parameter]):
        if interval[0] < value <= interval[1]:
            return interval_index * 0.25
    return None

# Применение нормализации к данным
normalized_data = np.zeros_like(data, dtype=float)  # Создаем пустой массив для нормализованных данных

for i in range(data.shape[1]):  # Проходим по столбцам данных
    for j in range(data.shape[0]):  # Проходим по строкам данных
        normalized_data[j, i] = normalize(i, data[j, i])  # Применяем нормализацию к каждому значению

print("Нормализованные данные:")
print(normalized_data)

print("Выходные данные:")
print(output_data)

np.random.seed(1)

# случайно инициализируем, на скрытом слое 16 нейрона
syn0 =  2*np.random.random((4,16)) - 1
syn1 =  2*np.random.random((16,3)) - 1

for j in range(500):
    # проходим вперёд по слоям 0, 1 и 2
    l0 =normalized_data
    l1 = nonlin(np.dot(l0, syn0))
    l2 = nonlin(np.dot(l1, syn1))
    # как сильно мы ошиблись относительно нужной величины?
    l2_error = output_data - l2
    if (j %100==0) :
        print("Error:" + str(np.mean(np.abs(l2_error))))
    if (j ==499) :
        print("l2:" + str( l2))
    # в какую сторону нужно двигаться?
    # если мы были уверены в предсказании, то сильно менять его не надо
    l2_delta = l2_error * nonlin(l2, deriv=True)
    # как сильно значения l1 влияют на ошибки в l2?
    l1_error = l2_delta.dot(syn1.T)

    # в каком направлении нужно двигаться, чтобы прийти к l1?
    # если мы были уверены в предсказании, то сильно менять его не надо
    l1_delta = l1_error * nonlin(l1, deriv=True)

    syn1 += l1.T.dot(l2_delta)
    syn0 += l0.T.dot(l1_delta)

def classify_cars_with_confidence(l2):
    results = []
    for row in l2:
        # Вычисление суммы вероятностей в строке
        row_sum = np.sum(row)
        # Вычисление процентов для каждого класса
        confidences = [value / row_sum * 100 for value in row]
        # Форматирование результатов
        result_str = ", ".join([f"{conf:.2f}%" for conf in confidences])
        results.append(result_str)
    return results

# Вызов функции и вывод результатов
results = classify_cars_with_confidence(l2)
print("Результаты классификации с уверенностями(Легковой автомобиль; Пассажирский транспорт; Грузовой транспорт):")
for result in results:
    print(result)


````
